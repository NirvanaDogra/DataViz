```python
import pandas as pd
data = pd.read_csv("/kaggle/input/dataset-for-apriori-analysis/guitar_apriori_analysis.csv")
```

```python
data = data.drop(['products'], axis=1)
```

```python

minSup = 15/data.shape[0]
print(minSup)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd

frequent_itemsets = apriori(data, min_support=minSup, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=minSup)


# Calculate Kulczynski Metric for each rule
def calculate_kulczynski(rule):
    precision = rule['support'] / rule['antecedent support']
    recall = rule['support'] / rule['consequent support']
    kulczynski = (precision + recall) / 2
    return kulczynski

rules['Kulczynski'] = rules.apply(calculate_kulczynski, axis=1)

# Calculate Imbalance Ratio for each rule
def calculate_imbalance_ratio(rule):
    support_a = rule['antecedent support']
    support_b = rule['consequent support']
    support_a_b = rule['support']
    imbalance_ratio = abs(support_a-support_b) / (support_a + support_b)
    return imbalance_ratio

rules['Imbalance Ratio'] = rules.apply(calculate_imbalance_ratio, axis=1)

# Print the association rules along with Kulczynski Metric and Imbalance Ratio
print(frequent_itemsets)
print("Association Rules with Metrics:")
print(rules)

print("\nAssociation Rules:")
print(rules)
```
