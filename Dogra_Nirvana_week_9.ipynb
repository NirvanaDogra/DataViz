```python
import plotly.express as px
import pandas as pd
import numpy as np

# Create sample data
np.random.seed(123)
n = 100
df = pd.DataFrame({'x': np.random.randn(n),
                   'y': np.random.randn(n),
                   'z': np.random.randint(1, 4, n)})

# Create scatter plot with trend line and color-coded markers
fig = px.scatter(df, x='x', y='y', trendline='ols', color='z',
                 color_discrete_sequence=['red', 'green', 'blue'])

# Update layout
fig.update_layout(title='Random Visualization',
                  xaxis_title='X Axis',
                  yaxis_title='Y Axis')

# Show figure
fig.show()
```
