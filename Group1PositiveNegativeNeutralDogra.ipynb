```python
from numpy.ma.core import negative
import matplotlib.pyplot as plt
```

```python
import pandas as pd

data = pd.read_csv("./store_data.csv", encoding="windows-1252")
```

```python
data.head()
```

Negative

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 10))
# Filter the data to include only negative profits
negativeProfit = data[data['Profit'] < 0]

# Create the bar plot
plt.bar(negativeProfit['State'], negativeProfit['Profit'], color = ['red'])

# Add axis labels and a title
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Losses')
plt.title('Losses by State')

```

Positive

```python
import matplotlib.pyplot as plt

# Filter the data to include only negative profits
positveProfit = data[data['Profit'] >= 0]


fig = plt.figure(figsize=(10, 10))
# Create the bar plot
plt.bar(positveProfit['State'], positveProfit['Profit'], color = ['green'])

# Add axis labels and a title
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Profits')
plt.title('Pofits by State')
```

Neutral 

```python
fig = plt.figure(figsize=(10, 10))
plt.bar(positveProfit['State'], positveProfit['Profit'], color = ['green'])
plt.bar(negativeProfit['State'], negativeProfit['Profit'], color = ['red'])
plt.xticks(rotation=90)
plt.show()
```
